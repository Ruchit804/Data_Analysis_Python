- Sort the values from the most to less expensive

```python
chipo['item_price'] = chipo['item_price'].astype(str).str.replace('$', '')

chipo['item_price'] = pd.to_numeric(chipo['item_price'])

chipo['price_per_item'] = chipo['item_price'] / chipo['quantity']

chipo_unique = chipo.drop_duplicates(subset=['item_name', 'quantity'])

item_prices = chipo_unique[['item_name', 'price_per_item']]

sorted_item_prices = item_prices.sort_values(by='price_per_item', ascending=False)

print(sorted_item_prices)
```

### 49. Write a Python script to display the
- a. Current date and time
- b. Current year
- c. Month of year
- d. Week number of the year
- e. Weekday of the week
- f. Day of year
- g. Day of the month
- h. Day of week

```python
import datetime

# Current date and time
now = datetime.datetime.now()
print("Current date and time:", now)

# Current year
print("Current year:", now.year)

# Month of year
print("Month of year:", now.month)

# Week number of the year
print("Week number of the year:", now.strftime("%U"))

# Weekday of the week (0=Monday, 6=Sunday)
print("Weekday of the week:", now.weekday())

# Day of year
print("Day of year:", now.strftime("%j"))

# Day of the month
print("Day of the month:", now.day)

# Day of week (0=Monday, 6=Sunday)
print("Day of week:", now.strftime("%w"))
```

### 50. Write a Python program to convert a string to datetime.
Sample String : Jul 1 2014 2:43PM 

Expected Output : 2014-07-01 14:43:00

```python
from datetime import datetime

# Sample string
date_string = "Jul 1 2014 2:43PM"

# Convert string to datetime
date_time_obj = datetime.strptime(date_string, "%b %d %Y %I:%M%p")

# Print the converted datetime object
print("Converted datetime:", date_time_obj)
```

### 51. Write a Python program to subtract five days from current date.

Current Date : 2015-06-22

5 days before Current Date : 2015-06-17

```python
from datetime import datetime, timedelta

# Current date
current_date = datetime.strptime("2015-06-22", "%Y-%m-%d")

# Subtract five days
new_date = current_date - timedelta(days=5)

# Print the result
print("Current Date:", current_date.strftime("%Y-%m-%d"))
print("5 days before Current Date:", new_date.strftime("%Y-%m-%d"))
```

### 52. Write a Python program to convert unix timestamp string to readable date.

Sample Unix timestamp string : 1284105682
    
Expected Output : 2010-09-10 13:31:22

```python
import datetime

# Unix timestamp string
timestamp_str = "1284105682"

# Convert Unix timestamp to datetime
timestamp = datetime.datetime.fromtimestamp(int(timestamp_str))

# Print the converted datetime
print("Converted datetime:", timestamp.strftime('%Y-%m-%d %H:%M:%S'))
```

### 54. Write a Python program to get days between two dates. 

```python
from datetime import datetime


date_str1 = '2023-05-10'
date_str2 = '2023-05-05'

date1 = datetime.strptime(date_str1, '%Y-%m-%d')
date2 = datetime.strptime(date_str2, '%Y-%m-%d')

delta = date1 - date2

print("Number of days between the two dates:", abs(delta.days))
```
